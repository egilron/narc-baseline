```python
from os import path
dummyfile = "bokmaal_aftenposten_06"
dummyfile = "nynorsk_vtbnn~20111117-5403"
ext = ".ann"
file_path = path.join("../../annotations", dummyfile + ext)
data = None
with open(file_path, "r", encoding="utf-8") as f:
    data = f.readlines()
    
```

```python
from conversion.ann_transform import get_references
```

```python
markables, references = get_references(data)
print(markables)
print(references)
markables
```

```python
def make_markable_mapping(markables, references):
    markable_coref_map = {}
    for mark in markables.keys():
        mark_refs = []
        for coref in references:
            if mark in coref:
                mark_refs.extend([_ref for _ref in coref if _ref != mark])
        if len(mark_refs) > 0:
            markable_coref_map[mark] = mark_refs

    return markable_coref_map

mapping = make_markable_mapping(markables, references)
mapping
```

```python
def expand_child(mapping, visited, entity, group):
    if entity not in visited:
        group.add(entity)
        visited.add(entity)
        print("visited", entity)
        print("visiting children:", mapping[entity])
        for child in mapping[entity]:
            expand_child(mapping, visited, child, group)

visited = set()
entity_groups = []
for entity in mapping:
    print("On entity: ", entity)
    entity_group = set()
    expand_child(mapping, visited, entity, entity_group)
    if len(entity_group) > 0:
        children = sorted(list(entity_group), key=lambda x: int(x.split("T")[-1]))
        entity_groups.append(children)
        print("children:", children)
    print("_"*30)
entity_groups
```

```python
from utils import mark_to_idx

idxs = []
for group in entity_groups:
    idxs.append(tuple(mark_to_idx(markables, markable) for markable in group))
len(idxs)
```
