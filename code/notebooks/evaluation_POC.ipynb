```python
import os

from batch_conll2jsonlines import convert

experiment_folder = "../../experiments/tomls/POC5/conll_logs"
output_folder = "../../converted_data/POC5"

for conll_file in os.listdir(experiment_folder):
    print(conll_file)
    path = os.path.join(experiment_folder, conll_file)
    file_id = conll_file.rsplit(".", 1)[0]
    print(file_id)
    convert(infpath=path, outfpath=os.path.join(output_folder, file_id + ".jsonl")) 
```

```python
import jsonlines
from corefeval import get_metrics

models = ["norBERT2", "xlm-r"]
for model in models:
    for split in ["dev", "test"]:
        gold = f"POC5-{model}_{split}_e20.gold.jsonl"
        pred = f"POC5-{model}_{split}_e20.pred.jsonl"

        with jsonlines.open(os.path.join(output_folder, gold)) as jsonl:
            gold_clusters = jsonl.read()["clusters"]
        with jsonlines.open(os.path.join(output_folder, pred)) as jsonl:
            pred_clusters = jsonl.read()["clusters"]
        
        conll_f1, metrics = get_metrics(
            predicted=pred_clusters,
            gold=gold_clusters,
            verbose=False,  # print detailed info
            modelname="None",
            dataset="Dummy file"
        )
        print(f"Model={model}, Split={split}")
        print(conll_f1)
        print(metrics)
```
