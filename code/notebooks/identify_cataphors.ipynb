```python
import os
import jsonlines

def flatten(xss):
    return [x for xs in xss for x in xs]

pronomen = "jeg eg meg du deg han hun ho hen den det ham han henne hen den det vi me oss dere de dokker dere dykk dei dem".split()
pronomen = "han ham hun henne"

root = "../../annotations_jsonlines"
for file_id in os.listdir(root):
    with jsonlines.open(os.path.join(root, file_id), "r") as reader:
        for obj in reader:  # it's always 1 object =)
            clusters = obj["clusters"]
            toks = flatten(obj["sentences"])
            for cluster in clusters:
                sorted_cl = sorted(cluster, key=lambda x: x[0])

                cl_toks = [toks[x:y+1] for (x, y) in sorted_cl]
                # print(cl_toks)
                
                first_mention = cl_toks[0]
                if len(first_mention) == 1:
                    if first_mention[0] in pronomen:
                        print(first_mention)
                        print(cl_toks)
                        print(sorted_cl)
                        print(file_id)
                        print("*****")
                    
```
